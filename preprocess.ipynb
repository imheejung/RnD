# Concatenating and preprocessing

```python
data = pd.merge(project,piv_pat, how='outer',on='id')
data = pd.merge(data,piv_pub, how='outer',on='id')
data = pd.merge(data,piv_roy, how='outer',on='id')
data = pd.merge(data,piv_sales, how='outer',on='id')
```

```python
data.shape
```

```python
# target only general R&D projects
data = data[data['division']=='일반연구개발사업']
data.shape
```

```python
# eliminat projects without government contributions
print("before eliminating projects: ", data.shape)

data = data.dropna(subset=['gov_funds'])

data = data[~(data['gov_funds']==0)]
print("after eliminating projects : ", data.shape)
```

```python
# normalize by the number of government funds
data['app_nm'] = data['app']/data['gov_funds']*100000000
data['reg_nm'] = data['reg']/data['gov_funds']*100000000
data['pub_nm'] = data['pub']/data['gov_funds']*100000000
data['royalty_nm'] = data['royalty']/data['gov_funds']
data['sales_nm'] = data['sales']/data['gov_funds']
data
```

```python
# make new data of number of registered patents/applied patents
data['pat_nm'] = data['reg_nm']/data['app_nm']
```

```python
# excluded projects with no research output
data = data[(data['app_nm'].notna()) | (data['reg_nm'].notna()) | (data['pub_nm'].notna()) | (data['royalty_nm'].notna()) | (data['sales_nm'].notna())]
```

```python
# change from date to year
data['start_date']= pd.to_datetime(data['start_date']) 
data['start_year'] = data['start_date'].dt.year
```

```python
print(np.sort(data['start_year'].unique()))
```

```python
data['end_date']= pd.to_datetime(data['end_date'], errors = 'coerce') 
data['end_year'] = data['end_date'].dt.year
```
